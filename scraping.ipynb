```python
import requests

from bs4 import BeautifulSoup

from pprint import pprint
```

```python
url = "https://www.nodc.noaa.gov/archive/arc0021/0002199/1.1/data/0-data/HTML/WMO-CODE/WMO4677.HTM"

html = requests.get(url)
html
```

```python
soup = BeautifulSoup(html.text, "html.parser")
```

```python
table = soup.find_all("table")

selection = table[1::2]

liste_td = []

for td in selection:
     X = td.find_all("td")
     liste_td.append(X)

new_liste_td = [[td for td in ligne if not td.has_attr("rowspan")]for ligne in liste_td]
    
cles = []
valeur = []

for bloc in range(len(new_liste_td)):
     cles.append(new_liste_td[bloc][0::2])
     valeur.append(new_liste_td[bloc][1::2])

clean_cles = []
clean_valeur = []

for infos in cles:
     for td in infos:
          clean_cles.append(td.get_text(strip=True))

for infos in valeur:
     for td in infos:
          clean_valeur.append(td.get_text(strip=True))

code_val = {}

for i in range(len(clean_cles)):
     code_val[clean_cles[i]] = clean_valeur[i]

pprint(code_val)
```
